```python
# Age Summary Statistics
df["age"].describe()
```

```python
# Feature engineer: Age Groups
bins = [18, 29, 39, 49, 59, 69, 79, 89, 95]  # Age ranges
labels = ["18-29", "30-39", "40-49", "50-59", "60-69", "70-79","80-89","90+"]  # Age group labels
df["age group"] = pd.cut(df["age"], bins=bins, labels=labels, right=True, include_lowest=True)
df.head()
```

```python
# Age Group Proportion
age_group_count = df["age group"].value_counts().sort_index()
age_group_percentage = (age_group_count / len(df)) * 100

age_group_summary = pd.DataFrame({
    "count": age_group_count,
    "percentage": age_group_percentage
})

print(age_group_summary)
```
